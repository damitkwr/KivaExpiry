```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import math
import datetime
```

```python
file = pd.read_csv("output_data/for_status_graphs.csv")
```

```python
file.columns
```

```python
plt.style.use('ggplot')
plt.rcParams.update({'font.size': 24})
plt.pyplot.figure(figsize = (12,9))
#plt.pyplot.grid()
plt.pyplot.tick_params(axis='both', which='major', labelsize=18)
plt.pyplot.tick_params(axis='both', which='minor', labelsize=18)
a = file.groupby(["activity", "status"]).apply(lambda df : df.shape[0]).unstack().reset_index()
a['percent'] = a['expired']/a['funded']
a = a.sort_values(by = 'percent')
g = sns.barplot(x = 'activity', y = 'percent', data = a, color="b")
g.set(xlabel='Activity', ylabel='% Expired')
g.set_xticklabels(labels = a['activity'],rotation=90)
plt.pyplot.title("Activities by What Percent of Loans Expired")

plt.pyplot.tight_layout(pad = 1.15)

g.get_figure().savefig("graphs/Activity_output.png")
```
